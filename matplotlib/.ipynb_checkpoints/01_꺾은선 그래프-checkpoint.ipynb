```python
import matplotlib.pyplot as plt
import matplotlib
```

#### 폰트 설정

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist]
```

#### 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]

plt.plot(x,y, label='범례', color='red', marker='o', markersize=10, markeredgecolor='red', markerfacecolor='yellow',
        linewidth=2, linestyle='None')
plt.legend(loc='upper right')

plt.title("꺾은선 그래프")
plt.xlabel('x축', color='red', loc='right') # left, center, right
plt.ylabel('y축', color='#00aa00', loc='top') # top, center, bottom
plt.xticks([1,2,3])
plt.yticks([2,4,6,8])

plt.show()
```

#### 포멧

```python
plt.figure(figsize=(10, 5), dpi=50) # dpi는 dots per inch, 확대
plt.plot(x, y, 'ro--', alpha=0.3) # color, marker, linestyle

for idx, txt in enumerate(y):
    plt.text(x[idx], y[idx]+0.3, txt, ha='center', color='blue') # ha는 horizontal alignment

plt.grid(axis='x', color='purple', alpha=0.5, ls='--', linewidth=2)
# plt.savefig('graph.png', dpi=100)
```

#### 여러 데이터

```python
days = [1, 2, 3] # 1일, 2일, 3일
az = [2, 4, 8] # (단위 : 만명) 1일부터 3일까지 아스트라제네카 접종인구
pfizer = [5, 1, 3]
moderna = [1, 2, 5]
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker='s', linestyle='-.')

plt.legend(ncol=3)
```
