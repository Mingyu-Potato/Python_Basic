## 막대그래프 기본

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['감자', '고구마', '연근']
values = [10, 8, 3]
colors=['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5, width=0.5)
plt.ylim(0, 15)
plt.xticks(rotation=45)
```

```python
labels = ['감자', '고구마', '연근']
values = [10, 8, 3]
xticks = ['1번 구황작물', '2번 구황작물', '3번 구황작물']

plt.bar(labels, values)
plt.xticks(labels, xticks)
```

## 막대그래프 심화

```python
labels = ['감자', '고구마', '연근']
values = [10, 8, 3]

plt.barh(labels, values)
plt.xlim(2,15)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar = plt.bar(labels, values)
plt.ylim(2, 12)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='red')
```

## 누적 막대 그래프

```python
name = ['채치수', '정대민', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협']
korean = [90,40,80,40,15,80,55,100]
english = [85,35,75,60,20,100,65,85]
math = [100,50,70,70,10,95,45,90]
```

```python
plt.bar(name, korean, label='국어')
plt.bar(name, english, bottom=korean, label='영어')
plt.bar(name, math, bottom=[k+e for k,e in zip(korean, english)], label='수학')

plt.xticks(rotation=60)
plt.legend()
plt.show()
```

## 다중 막대 그래프

```python
import numpy as np
```

```python
index = np.arange(8)
index
```

```python
plt.figure(figsize=(10,5))
plt.title('학생별 성적')

w=0.25
plt.bar(index-w, korean, width=w, label='국어')
plt.bar(index, english, width=w, label='영어')
plt.bar(index+w, math, width=w, label='수학')

plt.legend(ncol=3)
plt.xticks(index, name, rotation=60)
plt.show()
```
